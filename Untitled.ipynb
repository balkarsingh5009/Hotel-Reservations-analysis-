```python
# fix incorrect date
data["str_date"] = data.apply(lambda x: f'{x.arrival_year}-{x.arrival_month}-{x.arrival_date}', axis=1)
data["str_date"] = data["str_date"].replace("2018-2-29", "2018-2-28")
data["arr_date"] = pd.to_datetime(data["str_date"])
```

```python
data["test__arr_day_of_week"] = data.apply(lambda x: x.arr_date - datetime.timedelta(x.lead_time), axis=1).dt.weekday
data["is_weekend_lead_time"] = data["test__arr_day_of_week"].apply(lambda x: 1 if x > 5 else 0)
data["is_weekend_lead_time"].value_counts()
```
